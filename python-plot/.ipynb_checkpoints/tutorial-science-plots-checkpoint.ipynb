```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use(['science','ieee'])
```

```python
def model(x, p):
    return x ** (2 * p + 1) / (1 + x ** (2 * p))

x = np.linspace(0.75, 1.25, 10)
```

```python
fig, ax = plt.subplots()
for p in [10, 15, 20, 30, 50, 100]:
    ax.plot(x, model(x, p), label = p, marker='o', markersize=3)

ax.legend(title='Order')
ax.set(xlabel='Voltage (mV)')
ax.set(ylabel='Current ($\mu$A)')
ax.autoscale(tight=True)
fig.savefig('fig1.pdf')
fig.savefig('fig1.jpg', dpi=300)
```

```python
# Input data; groupwise
green_data = [16, 23, 22, 21, 13, 11, 18, 15]
blue_data =  [ 3,  3,  0,  0,  5,  5,  3,  3]
red_data =   [ 6,  6,  6,  0,  0,  0,  0,  0]
black_data = [25, 32, 28, 21, 18, 16, 21, 18]

labels = ['XI', 'XII', 'XIII', 'XIV', 'XV', 'XVI', 'XVII', 'XVIII']

# Setting the positions and width for the bars
pos = list(range(len(green_data))) 
width = 0.15 # the width of a bar
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(10,6))

bar1=plt.bar(pos, green_data, width,
                 alpha=1,
                 #color='w',
                 hatch='x', # this one defines the fill pattern
                 label=labels[0],
                 edgecolor='k', linewidth=2)

plt.bar([p + width for p in pos], blue_data, width,
                 alpha=0.5,
                 #color='w',
                 hatch='o',
                 label=labels[1],
                 edgecolor='k', linewidth=2)
    
plt.bar([p + width*2 for p in pos], red_data, width,
                 alpha=0.5,
                 #color='k',
                 hatch='',
                 label=labels[2],
                 edgecolor='k', linewidth=2)

plt.bar([p + width*3 for p in pos], black_data, width,
                 alpha=0.5,
                 #color='w',
                 hatch='*',
                 label=labels[3],
                 edgecolor='k', linewidth=2)


# Setting axis labels and ticks
ax.set_ylabel('Number of Switching')
ax.set_xlabel('Strategy')
ax.set_title('Grouped bar plot')
ax.set_xticks([p + 1.5 * width for p in pos])
ax.set_xticklabels(labels)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*5)
plt.ylim([0, max(green_data + blue_data + red_data) * 1.5])

# Adding the legend and showing the plot
plt.legend(['OLTC', 'SVC', 'SC', 'OLTC+SC+SVC'], loc='upper right')
plt.grid()
plt.show()
```

```python
patterns = ('-', '+', 'x', '\\', '*', 'o', 'O', '.')

fig = plt.gca()

# input data
mean_values = range(1, len(patterns)+1)

# plot bars
x_pos = list(range(len(mean_values)))
bars = plt.bar(x_pos, 
               mean_values, 
               align='center', 
               color='white',
               alpha=1, edgecolor='k', linewidth=2
               )

# set patterns
for bar, pattern in zip(bars, patterns):
     bar.set_hatch(pattern)
        

# set axes labels and formatting
fig.axes.get_yaxis().set_visible(False)        
plt.ylim([0, max(mean_values) * 1.1])
plt.xticks(x_pos, patterns)

plt.show()
```
